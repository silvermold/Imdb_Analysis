```python
from bs4 import BeautifulSoup as bs
import requests
```

# 1. Retrieve WebPage Data

In the code that follows, I will:

* Extract HTML from webpage
* Filter out unnecesarry information

```python
URL='https://www.imdb.com/chart/moviemeter/?sort=ir,desc&mode=simple&page=1'
req=requests.get(URL)
page=bs(req.content, 'html.parser')
```

```python
print(page.contents)
```

```python
target= page.find('tbody', class_='lister-list')
print(target.prettify())
```

```python
titles=target.find_all('td', class_='titleColumn')
```

```python
for entry in titles:
    name=entry.find('a')
    year=entry.find('span', class_='secondaryInfo')
    rating=entry.parent.find('td', class_='imdbRating')
    print(name.text.strip())
    print(year.text)
    print(f'Rating:{rating.text.strip()}')
    print()
```

# 2. Transform Data Into Useable Format

In the code that follows, I will:

* Format data for processing
* Clean the newly created dataframe

```python
import pandas as pd
```

```python
movies=pd.DataFrame(columns=['title','year','rating'], index=range(100))
movies
```

```python
movie_titles=[]
movie_years=[]
movie_ratings=[]
```

```python
for entry in titles:
    name=entry.find('a')
    year=entry.find('span', class_='secondaryInfo')
    rating=entry.parent.find('td', class_='imdbRating')
    
    movie_titles.append(name.text.strip())
    movie_years.append(year.text.strip())
    movie_ratings.append(rating.text.strip())
```

```python
movie_titles=pd.Series(movie_titles)
movie_ratings=pd.Series(movie_ratings)
movie_years=pd.Series(movie_years)
```

```python
movies['title']=movie_titles
movies['year']=movie_years
movies['rating']=movie_ratings
```

```python
movies
```

```python
movies=movies.drop(index=movies.index[88:100])
```

```python
movies
```

```python
movies['rating']=movies['rating'].astype(float)
```

```python
movies.info()
```

```python
N=0
```

```python
for year in movies['year']:
    movies['year'].loc[N]=movies['year'].loc[N][1:5]
    N+=1
```

```python
movies['year']=pd.to_datetime(movies['year'])
```
